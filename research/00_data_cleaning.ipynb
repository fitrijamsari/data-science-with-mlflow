# DATA CLEANING

## Rename Column

```python
def rename_columns(data, rename_dict):
    data.rename(columns=rename_dict, inplace=True)
    return data

# Dictionary for renaming columns
columns_to_rename = {
    'description': 'Description',
    '# of Floors': 'Num of Floors',
    'price': 'Price',
    'Property Size': 'Property Size in sqft'
}

# Apply the function to rename columns
df = rename_columns(df, columns_to_rename)
```

```python
df.columns
```

# Drop columns

```python
def drop_columns(data, columns):
    data.drop(columns=columns, inplace=True)
    return data

# List of columns to drop
columns_to_drop = ['Description', 'Ad List', 'Category', 'Facilities', 'Floor Range', 'Tenure Type', 'Building Name', 'Developer', 'Property Type', 'Firm Type', 'Firm Number', 'REN Number', 'Nearby Railway Station']

# Apply the function to drop specified columns
df = drop_columns(df, columns_to_drop)
```

# Clean Property Size

```python
# remove sq.ft. in Property size columna and convert to float64
def clean_and_convert_sqft_column(data, column_name):
    data[column_name] = data[column_name].apply(lambda s: s.replace('sq.ft.', '').strip()).astype("float64")
    return data

df = clean_and_convert_sqft_column(df, 'Property Size in sqft')
```

# Convert to Numeric

```python
def convert_to_numeric(data, columns):
    for col in columns:
        data[col] = data[col].notnull().astype('int64')
    return data

# List of columns to convert to numerical
columns_to_convert = ['Nearby School', 'Nearby Mall', 'Bus Stop', 'Mall', 'Park', 'School', 'Hospital', 'Highway','Railway Station']

# Apply the function to convert specified columns to numerical
df = convert_to_numeric(df, columns_to_convert)
```

```python
import numpy as np 

malaysia_states = [
    'Johor', 'Kedah', 'Kelantan', 'Melaka', 'Negeri Sembilan', 
    'Pahang', 'Perak', 'Perlis', 'Penang', 'Sabah', 'Sarawak', 
    'Selangor', 'Terengganu', 'Kuala Lumpur', 'Labuan', 'Putrajaya'
]

def assign_state_from_address(data, address_column, state_list):
    data['State'] = data[address_column].apply(lambda address: next((state for state in state_list if state in str(address)), np.nan))
    data.drop([address_column], axis=1, inplace=True)
    return data

df = assign_state_from_address(df, 'Address', malaysia_states)
```

# Replace Missing Values with Mean

```python
df['Total Units'].isnull().sum()
```

```python
def replace_missing_with_mean(data, columns):
    for col in columns:
        # Calculate mean and round to nearest integer
        mean_value = round(data[col].mean())
        data[col].fillna(mean_value, inplace=True)
        data[col] = data[col].astype(int)
        # data[col].fillna(data[col].mean(), inplace=True)
    return data


# List of columns with missing values to replace with mean
columns_with_missing_values = ['Total Units', 'Bedroom', 'Bathroom', 'Num of Floors', 'Parking Lot']

# Apply the function to replace missing values with mean in specified columns
df = replace_missing_with_mean(df, columns_with_missing_values)
```

```python
def replace_missing_random_year(data, columns):
    for col in columns:
        # Identify missing values
        missing_values = data[col].isnull()
        # Count the number of missing values
        num_missing_values = missing_values.sum()
        # Generate random years between 1985 and 2026
        random_years = np.random.randint(1985, 2027, size=num_missing_values)
        # Replace missing values randomly with generated years
        data.loc[missing_values, col] = random_years
        # Convert column to integers (optional)
        data[col] = data[col].astype(int)
    return data


df = replace_missing_random_year(df, ['Completion Year'])
```

```python
# def one_hot_encode(data, columns):
#     data_encoded = pd.get_dummies(data, columns=columns, prefix=columns, drop_first=True)

#     for col in data_encoded.columns:
#         if col.startswith(tuple(columns)):
#             data_encoded[col] = data_encoded[col].applymap(lambda x: 1 if x else 0)  # Convert boolean values to 0 or 1
#             data_encoded[col] = data_encoded[col].astype(int)
#     return data_encoded

def one_hot_encode(data, columns):
    data_encoded = pd.get_dummies(data, columns=columns, prefix=columns, drop_first=True)
    
    # Convert boolean values to 0 or 1 only for one-hot encoded columns
    for col in data_encoded.columns:
        if col.startswith(tuple(columns)):
            data_encoded[col] = data_encoded[col].astype(int)
    
    return data_encoded

# List of columns to perform one-hot encoding
columns_to_encode = ['Land Title', 'State']  # Replace with your column names

# Apply the function to perform one-hot encoding for multiple columns
df = one_hot_encode(df, columns_to_encode)
```

# Price

```python
def remove_currency_unit(data, columns):
    for col in columns:
        data[col] = data[col].apply(lambda s: "".join(s.split(" ")[1:])).astype('float64')
    return data

# List of columns with missing values to replace with mean
currency_column = ['Price']

# Apply the function to replace missing values with mean in specified columns
df = remove_currency_unit(df, currency_column)
```

```python
def shift_column_to_end(data, column):
    price_column = data.pop(column)  # Remove 'Price' column
    data[column] = price_column  # Add 'Price' column at the end
    return data

df = shift_column_to_end(df, 'Price')
```
